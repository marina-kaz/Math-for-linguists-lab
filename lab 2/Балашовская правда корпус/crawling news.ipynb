```python
import requests as r
from bs4 import BeautifulSoup
import regex as re
import time as t
```

```python
unavailable = []
# сюда будем класть юрл недоступных сайтов 
# и инфу об их статусе
```

```python
start = t.time()
'''чтобы сэкономить время, компилируем
регулярное выражение еще до цикла, так как 
оно все равно для всех новостей одинаковое'''
pattern_title = re.compile('(?=>).*(?=\|)')
pattern_source = re.compile('(?=\|).*(?=</title>)')
for i in range(3000): # последняя новость на сайте имеет индекс 2614, но возьмем с запасом
    '''генерируем адрес и делаем запрос'''
    url = 'http://balpravda.ru/node/' + str(i)
    response = r.get(url)
    '''в случае, если эта страница доступна, начинаем искать данные'''
    if not response.status_code == 200:
        unavailable.append([url, response.status_code])
    else:
        soup = BeautifulSoup(response.text) # получаем хтмл
        head_elem = soup.find_all("title")  # выделяем блок с названием
        title = re.findall(pattern_title, str(head_elem))[0][1:].strip() # само название
        source = re.findall(pattern_source, str(head_elem))[0][1:].strip() # название ресурса
        author = '' # балашовская правда не считает нужным подписывать автора статьи
        date = '' # и дату тоже 
        '''разделяем текст сайта по тексту заголовка
        последний элемент списка будет текстом статьи + нижняя панель'''
        pretext = soup.get_text().split(title)[2] 
        '''разделяем претекст по слову Акция, потому что текст нижней 
        панели начинается с этого слова. На всякий случай, если вдруг 
        в тексте статьи также встречается слово Акция, склеиваем по 
        нему же, оставив последний элемент списка (нижнюю панель) за рамками'''
        text = 'Акция'.join(pretext.split('Акция')[:-1]).strip() 
        file_name = str(i) + '.txt'
        path = 'Балашовская правда корпус/' + file_name
        with open(path, 'w', encoding='utf-8') as file:
            line = f'''URL====={url}
название источника====={source}
дата====={date}
автор====={author}
название статьи====={title}
текст статьи====={text}'''
            file.write(line)
finish = t.time()
finish-start
```
